````python
# Definir la ecuación de la recta.
x = [0, 8]
m = 0.5  # Pendiente.
b = 0.5  # Intercepción en y.

y = [m * i + b for i in x] # Puntos de la recta.

# Creao la gráfica.
fig, ax = plt.subplots()
ax.plot(x, y, color="blue", label="y = {}x + {}".format(m, b))

# Añado texto con la pendiente.
plt.text(0, m*5 + b - 1, "Pendiente: {}".format(m), fontsize=12,
         bbox=dict(facecolor="white", alpha=0.5))

# Agrego la línea punteada (1,1) a (7,1).
plt.plot([1, 7], [1, 1], linestyle="--", color="orange")
plt.text(4.02 ,0.65,"run", fontsize=12)

# Agregp la línea punteada (7,1) a (7,4).
plt.plot([7, 7], [1 ,4], linestyle="--", color="orange")
plt.text(7.35 ,2.65,"rise", fontsize=12)

# Puntos A y B
plt.scatter([1,7], [1 ,4], color="red")
plt.text(0.75 ,1.05,"A", fontsize=12)
plt.text(6.75 ,4.05,"B", fontsize=12)

# Etiquetas y título.
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Slope = rise / run")

# Rango del eje y.
plt.yticks(range(6))

# Cuadrícula en la gráfica.
plt.grid(True)

# Mostrar la gráfica resultante
plt.show()

print ("La pendiente es: ", b)
````
