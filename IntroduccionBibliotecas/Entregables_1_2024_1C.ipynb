````python
# Cargo los datos de cada archivo.
setosa = pd.read_csv("./1_datos/iris-setosa.csv")
versicolor = pd.read_csv("./1_datos/iris-versicolor.csv")
virginica = pd.read_csv("./1_datos/iris-virginica.csv")

# Reemplazo los valores numéricos de la columna 5 por los nombres de clase (índice 4).
setosa.iloc[:, 4] = "setosa"
versicolor.iloc[:, 4] = "versicolor"
virginica.iloc[:, 4] = "virginica"

# Concateno los datos de cada archivo en una sola variable.
iris_data = pd.concat([setosa, versicolor, virginica])

# Creo archivo CSV.
iris_data.to_csv("iris.csv", index=False)
````

2. Obtenga la media y desvio de cada atributo y compárelos con las mostradas en el archivo **iris.rst**, cree un dataframe con estos datos
